```python
# create a dictionary to tranform location name to number 
# feature engineering 
location_dict_new = {
    'Chennai': 7,
    'Noida': 6,
    'Bangalore': 5,
    'Hyderabad': 4,
    'Pune': 3,
    'Madurai': 2,
    'Lucknow': 1,
    'Other place': 0,
}
print(location_dict_new)
```

```python
# replace all the string with numbers 
def location(x):
  if str(x) in location_dict_new.keys():
    return location_dict_new[str(x)]
  else:
    return location_dict_new['Other place']

data_1 = attrdata["Location"].apply(location)
attrdata['New Location'] = data_1
attrdata.head()
```

```python
# Note that above we have created new column as "New Location"  and the old column "Location" still exist
```

```python
# get_dummies function convert the catigorycal data to binary(0,1) 
# hence increasing the number of column 
# this identify the feature is either true or false 
gen  = pd.get_dummies(attrdata['Function'])
gen.head()
```

```python
hr = pd.get_dummies(attrdata['Hiring Source'])
hr.head()
```

```python
# we know the value counts of Marital Status
Marital_dict
```

```python
# for Marital Status we have [Single, Marr, Div, NTBD,  Sep ]
# convert it to [Single, Marr, other status ]

def Mar(x):
  if str(x) in Marital_dict.keys() and Marital_dict[str(x)] > 100:
    return str(x)
  else:
    return 'other status'

data_1 = attrdata["Marital Status"].apply(Mar)
attrdata['New Marital'] = data_1
attrdata.head()
```

```python
New_Marital_dict = attrdata["New Marital"].value_counts()
New_Marital_dict
```

```python
# get_dummies function convert the catigorycal data to binary(0,1) 
# hence increasing the number of column 
# this identify the feature is either true or false 
Mr = pd.get_dummies(attrdata["New Marital"])
Mr.head()
```

```python
# Promoted/Non Promoted feature column has string Non Promoted and Promoted
# this can be converted to binary (0, 1)

def Promoted(x):
  if x == 'Promoted':
    return int(1)
  else:
    return int(0)

#creating new series with the data column
data_1 = attrdata["Promoted/Non Promoted"].apply(Promoted)
# adding this new feature to the data
attrdata['New Promotion'] = data_1
attrdata.head()
```

```python
# we know that Emp. Group have value count 
Emp_dict
```

```python
# adding other group 
Emp_dict['other group'] = 1
print(Emp_dict)
```

```python
# we will compress all the group 
Emp_dict_new = {
    'B1':4,
    'B2':3,
    'B3':2,
    'other group':1,
}

def emp(x):
  if str(x) in Emp_dict_new.keys():
    return str(x)
  else:
    return 'other group'


data_1 = attrdata['Emp. Group'].apply(emp)
attrdata['New EMP'] = data_1

# now New EMP cloumn has  [B1, B2, B3, other group]
# get_dummies function convert the catigorycal data to binary(0,1)
# and we are storing it in variable emp

emp = pd.get_dummies(attrdata['New EMP'])
attrdata.head()
```

```python
# Job Role
job_dict
```

```python
def Job(x):
  if x == 'Yes':
    return int(1)
  else:
    return int(0)

data_1 = attrdata['Job Role Match'].apply(Job)
attrdata['New Job Role Match'] = data_1
attrdata.head()
```

```python
dataset = pd.read_csv("drive/MyDrive/DataScienceProject/AttritionRate/processed table.csv")
#
dataset = pd.DataFrame(dataset)
#
y = dataset["Stay/Left"]
X = dataset.drop("Stay/Left", axis=1) 
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=4)
```

```python
#
```

```python
# let check the performance of all 
lr  = LogisticRegression(C = 0.1, random_state= 42, solver='liblinear') # applying hyperparameter
dt  = DecisionTreeClassifier()
rm  = RandomForestClassifier()
gnb = GaussianNB()
# knn = sklearn.svm.SVC
#supvm = sklearn.neighbors.KNeighborsClassifier
```

```python
for a, b in zip([lr, dt, rm, gnb], ["Logistic Regression", "Decision Tree", "Random Forest", "Naive Bayes"]):
  a.fit(X_train, y_train)

  # prediction with training data  
  prediction = a.predict(X_train)
  # prediction with text data 
  y_pred=a.predict(X_test)

  # checking accuracy with actual result for prediction on y_train for X_train data  
  score1 = accuracy_score(y_train,prediction)
  # checking accuracy with actual result for prediction on y_text for X_text data
  score = accuracy_score(y_test, y_pred)

  msg1 = "[%s] training data accuracy is : %f" % (b, score1)
  msg2 = "[%s] testing data accuracy is : %f" % (b, score)

  print(msg1)
  print(msg2)
```

```python
# Standardization and Normalization are data preprocessing techniques whereas Regularization is used to improve model performance
# we can use Standardization or Regularization
```

```python
model = lr.fit(X_train, y_train)
```
